## 파생변수 생성하기

```python
df.info()
```

```python
#연월일 파생변수 만들기
# df["사용월"] = pd.to_datetime(df["사용월"])  ==> 이건 처리가 안됨. 1970년도로 계속 나옴 int 타입을 string 포멧으로변경후 재변경

df['사용월'] = pd.to_datetime(df['사용월'].astype(str), format='%Y%m')
df['연'] = pd.to_datetime(df['사용월']).dt.year
df['월'] = pd.to_datetime(df['사용월']).dt.month
df
```

```python
# 승차인원, 하차인원 만들기
a = df[['연','월','호선명','지하철역']]
b = df.filter(regex='승차')
k = df.filter(regex='하차')

#ridetime = 시간대별 승차인원 / alightime = 시간대별 하차인원  
ridetime = a.join(b)
alighttime = a.join(k)
# ridetime.info()
# ridetime.columns = 
# ['04시-05시 승차인원','05시-06시 승차인원','06시-07시 승차인원','07시-08시 승차인원','08시-09시 승차인원','09시-10시 승차인원','10시-11시 승차인원',
# '11시-12시 승차인원','12시-13시 승차인원','13시-14시 승차인원','14시-15시 승차인원','15시-16시 승차인원','16시-17시 승차인원','17시-18시 승차인원','18시-19시 승차인원',
# '19시-20시 승차인원', '20시-21시 승차인원', '21시-22시 승차인원', '22시-23시 승차인원','23시-24시 승차인원', '00시-01시 승차인원', '01시-02시 승차인원', '02시-03시 승차인원',
# '03시-04시 승차인원']

df['승차인원']= ridetime.iloc[:, -25:-1].sum(axis=1)
df['하차인원']= alighttime.iloc[:, -25:-1].sum(axis=1)

df.head(1)
# # cols_to_sum = df.columns[ : df.shape[1]]
# df['ridetime_total']= df[ridetime.columns].sum(axis=1)
# df['ridetime_total']
# .iloc[:,-28:-1]

# df['ridetime_total']= ridetime.iloc[:, -18:-1].sum(axis=1)
# ridecount=pd.pivot_table(df, index = ["호선명","지하철역"], aggfunc = np.sum)
```

```python
#지하철별 하차인원 top 5 변수생성하기
#피봇 테이블로 하차인원 df 
data_station = df.pivot_table(index = ['지하철역'], values = '하차인원', aggfunc='sum').astype('int')
data_station
```

```python
alightcount = data_station.sort_values(by="하차인원",ascending=False).head(5).reset_index()
alightcount
```

```python
df_sub_station = alightcount[alightcount['하차인원']>=180000000]
df_sub_station
```

```python
list_df =df_sub_station['지하철역'].to_list()
list_df[4]
```

```python
#top5 지하철역 df_sub 변수생성
df_sub = pd.DataFrame()
for i in list_df:
     a =df[df['지하철역']==i]
     df_sub= df_sub.append(a)
df_sub 
```
